```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
ABC_company = pd.read_csv("/content/drive/MyDrive/ABC_Company.csv")
```

```python
ABC_company
```

```python
# Replacing 'height' column with random values between 150 and 180
ABC_company['height'] = np.random.randint(150, 181, ABC_company.shape[0])

# Checking for missing values
ABC_company.isnull().sum()
```

```python
# Employee count per team
team_distribution = ABC_company['Team'].value_counts()
team_distribution
```

```python
# Calculating percentage
team_percentage = (team_distribution / len(ABC_company)) * 100
team_percentage
```

#Outcome:
Ensures clean and consistent data for analysis.

## Visualization

```python
# # 1. Determine the distribution of employees across each team and calculate
# the percentage split relative to the total number of employees.
```

```python
## Visualization - BAR CHART
plt.figure(figsize=(10, 5))
sns.barplot(x=team_distribution.index, y=team_distribution.values, palette="Paired")
plt.xlabel("Teams")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Teams")
## rotates the x-axis tick labels by 85 degree to avoid overlapping.
plt.xticks(rotation=85)
plt.show()
```

## Outcome & Insights:


*   Identifies which teams have the most/least employees.
*   Highlights possible overstaffed or understaffed teams.



```python
## 2. Segregate employees based on their positions within the company.
```

```python
# Employee count per position
position_distribution = ABC_company['Position'].value_counts()
position_distribution
```

```python
## Visualization - PIE CHART
plt.figure(figsize=(8, 8))
plt.pie(position_distribution, labels=position_distribution.index, autopct="%1.1f%%", colors=sns.color_palette("muted"))
plt.title("Distribution of Employees by Position")
plt.show()
```

# Outcome & Insights:



*   Helps understand workforce distribution across job roles.
*   Identifies whether a company is top-heavy (too many managers) or bottom-heavy.








```python
## 3. Identify the predominant age group among employees.
```

```python
# Defining age groups
bins = [20, 30, 40, 50, 60]
labels = ['20-30', '30-40', '40-50', '50-60']
ABC_company['age_group'] = pd.cut(ABC_company['Age'], bins=bins, labels=labels, right=False)
```

```python
# Counting employees in each age group
age_group_distribution = ABC_company['age_group'].value_counts()
age_group_distribution
```

```python
## Visualization - HISTOGRAM
plt.figure(figsize=(8, 5))
sns.histplot(ABC_company['Age'], bins=10, kde=True, color='green')
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Age Distribution of Employees")
plt.show()
```

## Outcome & Insights:


*   Identifies the dominant workforce age range.
*   Highlights potential need for workforce planning (e.g., retirement risks).





```python
## 4. Discover which team and position have the highest salary expenditure.
```

```python
# Sum of salaries by team
team_salary_expense = ABC_company.groupby('Team')['Salary'].sum().sort_values(ascending=False)
```

```python
# Sum of salaries by position
position_salary_expense = ABC_company.groupby('Position')['Salary'].sum().sort_values(ascending=False)
```

```python
# Visualization - BAR CHART

plt.figure(figsize=(10, 5))
sns.barplot(x=team_salary_expense.index, y=team_salary_expense.values, palette="rocket")
plt.xlabel("Teams")
plt.ylabel("Total Salary Expense")
plt.title("Salary Expenditure by Team")
plt.xticks(rotation=85)
plt.show()
```

```python
## Visualization - LINE CHART
plt.figure(figsize=(8, 5))
sns.lineplot(x=position_salary_expense.index, y=position_salary_expense.values, marker="o", color="red", linewidth=2)
plt.xlabel("Position")
plt.ylabel("Total Salary Expense")
plt.title("Salary Expenditure by Position")
plt.show()
```

## Outcome & Insights:


*   Highlights the teams and positions with the highest financial impact.
*   Helps in budget allocation and salary adjustments.



```python
## 5. Investigate if there's any correlation between age and salary, and represent it visually.
```

```python
# Calculating correlation
correlation = ABC_company[['Age', 'Salary']].corr()
```

```python
## Visualization - HEATMAP
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='rocket', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Employee Data")
plt.show()
```

## Outcome & Insights:


*  A strong positive correlation suggests salaries increase with experience.
* A weak correlation indicates salaries may depend more on skills than age.
